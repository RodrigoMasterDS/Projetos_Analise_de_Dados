**O que é Renda Per Capita?** <p>
A renda per capita é um dos indicadores socioeconômicos que avaliam o grau de desenvolvimento econômico de um determinado lugar. A média é obtida através da divisão do Produto Nacional Bruto (PNB) pelo número total de habitantes.

```python
# Bibliotecas
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Lendo a base de dados
Base_Dados = pd.read_excel('Dados_Pib.xlsx')

# Verificando
Base_Dados.head(27)
```

```python
Base_Dados_2016 = Base_Dados[Base_Dados['Ano'] == 2016] \
    .sort_values(by='PIB per capita', ascending=True) \
    .drop_duplicates(subset='Territorialidades', keep='first') \
    .loc[:, ['Territorialidades', 'Ano', 'PIB per capita']]
Base_Dados_2016.head(28).reset_index(drop=True)
```

```python
# Sistema de Grids

# Cor Fundo de Fundo
Cor_Fundo = '#f5f5f5'

# Criar o sistema de Grids
Grid_Graficos = sns.FacetGrid( Base_Dados, col='Territorialidades', hue='Territorialidades', col_wrap=4 )

# Adicionar grafico linhas em cada gráfico
Grid_Graficos = Grid_Graficos.map( plt.plot, 'Ano', 'PIB per capita')

# Adiconar uma sombra + Ajuste do titulo
Grid_Graficos = Grid_Graficos.map( plt.fill_between,  'Ano', 'PIB per capita', alpha=0.2).set_titles('{col_name} Territorialidades')

# Filtrar o titulo
Grid_Graficos = Grid_Graficos.set_titles('{col_name}')

# Adicionar um subtitulo
Grid_Graficos = Grid_Graficos.fig.suptitle(
    'Evolução da Renda per capita por Estado' , fontsize=18
)

# Ajustando
plt.subplots_adjust( top=0.92 )
```

#### Podemos analisar mediante o gráfico acima que o PIB no período de 2013 a 2016,na maioria dos estados diminuiu de forma pouca expressiva, poucos estados apresentaram crescimento e ainda, pouco expressivo.

#### As regiões Sudeste e Sul, em geral, apresentaram níveis de renda per capita mais elevados e trajetórias de crescimento mais consistentes, em comparação com as regiões Norte e Nordeste.

#### Distrito Federal: Apresenta um destaque positivo, com uma renda per capita significativamente superior à média nacional e uma tendência de crescimento contínuo. Essa dinâmica pode estar relacionada à concentração de atividades governamentais e de serviços em Brasília.
