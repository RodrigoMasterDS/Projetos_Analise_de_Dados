Análise de Dados

**Sobre o conjunto de dados:** Dados do comportamento da bolsa de valores referente às ações da Magazine Luiza durante o ano de 2021, demonstrando valores min, max, abertura, fechamento, volume comercializado e ajuste de fechamento.

```python
# Importando as libs:

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go

import warnings
warnings.filterwarnings('ignore')
```

```python
# Lendo os dados:
Base_dados = pd.read_excel('Vase_004 - Magalu - Sem Resolução.xlsx')

# Verificando:
Base_dados.head()
```

```python
Base_dados.shape
```

```python
Base_dados.info()
```

```python
Base_dados.describe()
```

```python
# Séries Temporais / Colocando as Datas no local dos Index
Dados = Base_dados.set_index('Data')

Dados.head()
```

```python
plt.style.use('seaborn-v0_8-darkgrid')
plt.figure(figsize=(16, 5))
plt.title('Análise de Ações Magalu - Fechamento', fontsize=15, loc='left')
plt.plot(Dados.index, Dados['Fechamento'])

plt.xlabel('Período da Cotação')
plt.ylabel('Valor em R$');
```

```python
Dados.tail()
```

```python
# Tirando uma média móvel de 05 dias
Media_movel_05 = Dados['Fechamento'].rolling(5).mean()
Media_tendencia_30 = Dados['Fechamento'].rolling(30).mean()
```

```python
plt.style.use('seaborn-v0_8-darkgrid')
plt.figure(figsize=(16, 5))
plt.title('Análise de Ações Magalu - Fechamento', fontsize=15, loc='left')
plt.plot(Dados.index, Dados['Fechamento'])
plt.plot(Dados.index, Media_movel_05)
plt.plot(Dados.index, Media_tendencia_30)

plt.xlabel('Período da Cotação')
plt.ylabel('Valor em R$');
```

```python
# Boxplot Mensal

Base_dados['Mês'] = Base_dados['Data'].dt.month

plt.figure(figsize=(16, 5))
sns.boxplot(data=Base_dados, x='Mês', y='Fechamento', palette='viridis');
```

```python
Base_dados.groupby(by = ['Mês']).describe()['Fechamento']
```

```python
# Análise Gráfica utilizando Candles
Gráfico = go.Figure(
    data=[
        go.Candlestick(
            x=Dados.index,
            open=Dados['Abertura'],
            high=Dados['Maior'],
            low=Dados['Menor'],
            close=Dados['Fechamento']
        )
    ]
)
Gráfico.update_layout( xaxis_rangeslider_visible=False )

Gráfico.show()
```

### Diante desses dados analisados de mercado da bolsa de valores da ação da Magalu no ano de 2021, verificamos uma forte queda ao longo dos meses, aprensentando no mes de Março a maior variação entre abertura e fechamento, e em Junho a menor.
### Começou o mês de janeiro custando 25.26 e terminou em dezembro custando 7,22, demonstrando uma queda anual de 71,42%
