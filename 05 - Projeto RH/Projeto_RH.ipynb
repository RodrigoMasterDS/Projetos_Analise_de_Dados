```python
# Importando as Libs:

# Libs para Modelagem e Matrizez
import pandas as pd
import numpy as np

# Libs para análises gráficas
import matplotlib.pyplot as plt
import seaborn as sns

# Lib para ignorar avisos
import warnings

# Desabilitando avisos
warnings.filterwarnings('ignore')
```

```python
# Lendo base de dados:
Base_Dados = pd.read_csv('Salary_Data.csv')

# Verificando
Base_Dados.head()
```

```python
# Renomeando as colunas:
Base_Dados.rename(columns={
    'YearsExperience':'Anos_Experiência',
    'Salary':'Salário'}, inplace=True)

# Verificando
Base_Dados.head()
```

```python
Base_Dados.info()
```

```python
Base_Dados.shape
```

```python
# Campos nulos
Base_Dados.isnull().sum()
```

```python
# Grafico Nulos
sns.heatmap( Base_Dados.isnull(), cbar=False );
```

```python
Base_Dados.describe()
```

```python
plt.figure( figsize=(10, 6) )
plt.title('Análise do Salário')
sns.kdeplot( Base_Dados['Salário'], color='green' );
```

```python
plt.figure( figsize=(10, 6) )
plt.title('Análise da Experiência')
sns.kdeplot( Base_Dados['Anos_Experiência'], color='blue' );
```

```python
# Distribuição
plt.figure( figsize=(10, 5) )
sns.boxplot( Base_Dados['Salário'] );
```

```python
# Distribuição
plt.figure( figsize=(10, 5) )
sns.boxplot( Base_Dados['Anos_Experiência'] );
```

```python
# Distribuição
plt.figure( figsize=(10, 5) )
sns.scatterplot( data=Base_Dados, x='Salário', y='Anos_Experiência');
```

```python
plt.figure( figsize=(10, 5) )
sns.regplot( data=Base_Dados, x='Salário', y='Anos_Experiência');
```

```python
# Correlação
Correlacao = Base_Dados.corr('pearson')
plt.figure( figsize=(10, 5) )
sns.heatmap( Correlacao, annot=True, cbar=True );
```
